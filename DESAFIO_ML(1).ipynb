##Tratamento de dados

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/datasets/data_cancer2.csv', sep = ',', encoding = 'iso-8859-1')
df.drop(['Unnamed: 32'], axis = 1)
df.head()
```

```python
df2 = pd.DataFrame.copy(df)

df2['diagnosis'].replace({'M': 1, 'B': 0}, inplace = True)
df2.head()
```

```python
df3 = pd.DataFrame.copy(df2)
```

```python
df3.head()
```

```python
df3.replace({'[^0-9]': ''}, regex = True)
df3
```

```python
def ajustar_formato(x):

    if isinstance(x, str) and '.' not in x:

        return f'{int(x) / 100:.2f}'

    else:

        return x

df4 = df3.map(ajustar_formato)
df4.drop(['Unnamed: 32'], axis = 1)
df4.dropna
df4.head(3)
```

##Previsores

```python
previsores = df4.drop(['id', 'diagnosis', 'Unnamed: 32'], axis=1).values  # Pegar todas as colunas exceto 'id' e 'diagnosis'

print(previsores[:5])
```

##Alvo

```python
alvos = df4['diagnosis'].values

print(alvos[:5])
```

##Escalonamento

```python
df4.describe()
```

```python
from sklearn.preprocessing import StandardScaler

df4 = df.drop(['id', 'Unnamed: 32'], axis = 1)

scaler = StandardScaler()
previsores_esc = scaler.fit_transform(previsores)
previsores_esc
```

```python
previsores_df = pd.DataFrame(previsores_esc)
previsores_df.head()
```

##Base de treino e teste

```python
from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(previsores, alvos, test_size = 0.3, random_state = 0)
```

```python
x_treino.shape
```

##Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

naive = GaussianNB()
naive.fit(x_treino, y_treino)
```

#Métricas

```python
previsoes = naive.predict(x_teste)
print(previsoes)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

accuracy_score(y_teste, previsoes)
print('Acurácia: %.2f%%' % (accuracy_score(y_teste, previsoes)*100))
```

```python
confusion_matrix(y_teste, previsoes)
```

```python
print(classification_report(y_teste, previsoes))
```

#Análise de dados de treino

```python
previsoes_treino = naive.predict(x_treino)
print(previsoes_treino)
```
